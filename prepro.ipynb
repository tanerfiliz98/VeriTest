```python
from sklearn import preprocessing
lbe = preprocessing.LabelEncoder() 
```

```python
def labelEncNameMap(lbe,dataset):
    lbe.fit(dataset)
    lbe_name_mapping = dict(zip(lbe.classes_, lbe.transform(lbe.classes_)))
    print(lbe_name_mapping)
```

```python
labelEncNameMap(lbe,dataset["durumu"])
```

```python
datasetPre=pd.DataFrame()
datasetPre["marka"]=lbe.fit_transform(dataset['marka'])
datasetPre
```

```python
datasetPre["seri"]=lbe.fit_transform(dataset['seri'])
datasetPre
```

```python
datasetPre["model"]=lbe.fit_transform(dataset['model'])
datasetPre
```

```python
datasetPre["yil"]=dataset['yil']
datasetPre
```

```python
datasetPre["yakit"]=lbe.fit_transform(dataset['yakit'])
datasetPre
```

```python
datasetPre["vites"]=lbe.fit_transform(dataset['vites'])
datasetPre
```

```python
datasetPre["km"]=dataset['km']
datasetPre
```

```python
guc=[]
for x in dataset["motor gucu"]:
    if "-" not in x:
        guc.append(x.split()[0])
    else:
        guc.append((float(x.split()[0])+float(x.split()[2]))/2.0)
datasetPre["motor gucu"]=guc
datasetPre
```

```python
hacim=[]
for x in dataset["motor hacmi"]:
    if "-" not in x:
        hacim.append(x.split()[0])
    else:
        hacim.append((float(x.split()[0])+float(x.split()[2]))/2.0)
datasetPre["motor hacmi"]=hacim
datasetPre
```

```python
datasetPre["cekis"]=lbe.fit_transform(dataset['cekis'])
datasetPre
```

```python
datasetPre["kimden"]=lbe.fit_transform(dataset['kimden'])
datasetPre
```

```python
datasetPre["durumu"]=lbe.fit_transform(dataset['durumu'])
datasetPre
```

```python
datasetPre["fiyat"]=dataset["fiyat"].str.replace(" TL","")
datasetPre["fiyat"]=datasetPre["fiyat"].str.replace(".","")
datasetPre
```

```python
datasetPre.dtypes
```

```python
datasetPre = datasetPre.astype({"yil":int ,"km":int,"motor gucu":float,"motor hacmi":float,"fiyat":int})
```

```python
from openpyxl import load_workbook
book = load_workbook(r'C:\Users\taner\Desktop\MakineOgrenmesi1\veriSeti.xlsx')
writer = pd.ExcelWriter(r'C:\Users\taner\Desktop\MakineOgrenmesi1\veriSeti.xlsx', engine='openpyxl')
writer.book = book
datasetPre.to_excel(writer, sheet_name='Ön işleme sonrası veri',index = False)
writer.save()
```
